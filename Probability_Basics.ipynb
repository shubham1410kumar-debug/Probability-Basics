# **Probability Basics**

1. A die is rolled. What is the probability of getting:
(a) An even number
(b) A number greater than 4
  - (a) Even number:- {2,4,6} 3 Outcomes
        P(even)= 3/6 = 1/2
    (b) Number greater than 4:- {5,6} 2 outcomes
        P(X>4)= 2/6 = 1/3

2. In a class of 50 students:
20 like Mathematics (M)
15 like Science (S)
5 like both subjects
What is the probability that a student chosen at random likes Mathematics or Science?
  - Total Students=50
    M=20, S=15, Both=5, Students who like M or S= 20+15-5=30
    P(M or S)= 30/50= 3/5

3. A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?
  - Bag: 3 red, 2 blue → total 5 balls ​.
    First ball drawn is red and not replaced, so remaining: 2 red, 2 blue → total 4.
    P(next is red∣first red)=2/4=1/2

4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?
  - Use Stratified Sampling:                                       
    Divide population into strata (boys, girls).
    Select samples from each stratum to ensure required representation.So, method: Stratified Sampling, because it guarantees proportional/equal representation of both genders in the sample.

5. The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.
Find the sampling error.
  - Population mean = 160 cm.
    Sample mean (100 students) = 158 cm ​.
    Sampling error =∣sample mean−population mean∣
    =∣158−160∣=2 cm

6. The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?
  - Population mean salary = ₹50,000, population standard deviation σ=₹5,000
  Sample size n=100.
  Standard Error of Mean (SEM):
  SEM=5,000 $$\sqrt{100}$$  = 5,000/10

7.  In a group of 100 students:
40 like Cricket (C)
30 like Football (F)
10 like both Cricket and Football
Find the probability that a student likes at least one sport.
  - Total students = 100.
    C = 40, F = 30, both (C ∩ F) = 10
    Students who like at least one sport:
    ∣C∪F∣=∣C∣+∣F∣−∣C∩F∣=40+30−10=60 ​.
    P(at least one sport)=60/100
                         =3/5
8. From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?
  - Deck: 52 cards, 4 Aces
    First card Ace: 4/52
    Second card Ace (without replacement): 3/51
    P(both Aces)=4/52 X 3/51 = 12/2652
                = 1/221

9. A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?
  - Defective rate = 2%
    P(defective)=0.02.
    Non-defective probability = 1 − 0.02 = 0.98
    For 5 bulbs, all non-defective (independent):
    P(all non-defective) ≈0.9039

10. Differentiate between discrete and continuous random variables with examples.
  - Discrete random variable:
    Takes countable values (usually integers).
    Example: Number of heads in 5 coin tosses; number of customers entering a shop in an hour.
    Continuous random variable:
    Takes any value in an interval (uncountably infinite).
    Example: Height of students, time taken to finish a task, weight of a person.
